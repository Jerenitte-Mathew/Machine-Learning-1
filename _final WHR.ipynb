# <span style = "color: blue"> Model Selection and Training </span> 

##   Linear Regression

#### Import necessary libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

### Define features and target variables

```python
X = df[['Log GDP per capita', 'Social support', 'Healthy life expectancy', 
          'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
y = df['Happiness score']
```

### Perform train-test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

#### Predictions

```python
y_pred = lr_model.predict(X_test)
y_pred 
```

### Evaluation Matrics

```python
n = X_test.shape[0]  # Number of observations in test set
p = X_test.shape[1]  # Number of predictors
mse_lr = mean_squared_error(y_test, y_pred)
r2_lr = r2_score(y_test, y_pred)
mae_lr = mean_absolute_error(y_test, y_pred)
rmse_lr = math.sqrt(mse_lr)
adjusted_r2_lr = r2_lr - (1 - r2_lr) * (p / (n - p - 1))

print("Linear Regression Evaluation Metrics:")
print("Mean Squared Error (MSE):", mse_lr)
print("Root Mean Squared Error (RMSE):", rmse_lr)
print("Mean Absolute Error (MAE):", mae_lr)
print("R-squared (R2):", r2_lr)
print("Adjusted R-squared:", adjusted_r2_lr)
```

#### Scatterplot

```python
plt.figure(figsize=(8, 6))

# Plot y_test (Actual) in blue
plt.scatter(y_test, y_test, color='blue', label='Actual Happiness Score')

# Plot y_pred (Predicted) in red
plt.scatter(y_test, y_pred, color='red', label='Predicted Happiness Score')


plt.xlabel('Happiness Score')
plt.ylabel('Feature Value')
plt.title('Actual vs. Predicted Happiness Score')
plt.legend()
plt.show()
```

### Plot the distribution of actual and predicted values

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning, module='seaborn')
plt.figure(figsize=(10, 5))
sns.histplot(y_test, color='blue', label='Actual', kde=True, stat="density", linewidth=0)
sns.histplot(y_pred, color= 'green', label='Predicted (Linear)', kde=True, stat="density", linewidth=0)
plt.title('Distribution of Actual vs Predicted Values(LinearRegression)')
plt.xlabel('Happiness Score')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## Polynomial Regression

#### Import necessary libraries

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import math
```

#### Polynomial Regression model

```python
degree =2  # Set the degree of the polynomial
poly_features = PolynomialFeatures(degree=degree)
X_train_pr = poly_features.fit_transform(X_train)
X_test_pr = poly_features.transform(X_test)
```

```python
pr_model = LinearRegression()
```

```python
pr_model.fit(X_train_pr, y_train)
```

#### Prediction

```python
y_pred = pr_model.predict(X_test_pr)
y_pred 
```

### Evaluation metrics

```python
mse_pr = mean_squared_error(y_test, y_pred)
mae_pr = mean_absolute_error(y_test, y_pred)
r2_pr = r2_score(y_test, y_pred)
rmse_pr= math.sqrt(mse_pr)
adjusted_r2_pr = r2_pr - (1 - r2_pr) * (p / (n - p - 1))

print("Polynomial Regression Evaluation Metrics:")
print("Mean Squared Error (MSE):", mse_pr)
print("Root Mean Squared Error (RMSE):", rmse_pr)
print("Mean Absolute Error (MAE):", mae_pr)
print("R-squared (R2):", r2_pr)
print("Adjusted R-squared:", adjusted_r2_pr)
```

#### Scatterplot

```python
plt.figure(figsize=(8, 6))

# Plot y_test (Actual) in blue
plt.scatter(y_test, y_test, color='blue', label='Actual Happiness Score')

# Plot y_pred (Predicted) in red
plt.scatter(y_test, y_pred, color='red', label='Predicted Happiness Score')


plt.xlabel('Happiness Score')
plt.ylabel('Feature Value')
plt.title('Actual vs. Predicted Happiness Score')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(y_test, color='blue', label='Actual', kde=True, stat="density", linewidth=0)
sns.histplot(y_pred, color='green', label='Predicted (Polynomial)', kde=True, stat="density", linewidth=0)
plt.title('Distribution of Actual vs Predicted Values (Polynomial Regression)')
plt.xlabel('Happiness Score')
plt.ylabel('Density')
plt.legend()
plt.show()
```
